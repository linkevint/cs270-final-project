## Analyze Categories

```python
# remove businesses with no category
# only keep businesses that have 'Restaurants' as a category
business_categories_df = business_df[business_df['categories'].notnull()]
business_categories_df = business_categories_df[business_categories_df['categories'].str.contains('Restaurants')]
business_categories_df = business_categories_df[business_categories_df['categories'].str.contains('Food')]
```

```python
# filter down to business_id and categories
business_categories_df = business_categories_df[['business_id', 'categories']]
business_categories_df
```

```python
# get list of all possible categories
set(', '.join(business_categories_df['categories']).split(', '))
```

```python
dumplings = business_categories_df[business_categories_df['categories'].str.contains('Dumplings')]
for x in dumplings['categories']:
    if 'Chinese' not in x:
        print(x)
```

```python
dimsum = business_categories_df[business_categories_df['categories'].str.contains('Dim Sum')]
for x in dimsum['categories']:
    if 'Chinese' not in x:
        print(x)
```

```python
hakka = business_categories_df[business_categories_df['categories'].str.contains('Hakka')]
for x in hakka['categories']:
    print(x)
```

## Analyze Dietary Restrictions

```python
business_diet_df = business_df[business_df['attributes.DietaryRestrictions'].notnull()]
business_diet_df = business_diet_df[business_diet_df['attributes.DietaryRestrictions'] != 'None']
```

```python
# filter down to business_id and attributes.DietaryRestrictions
business_diet_df = business_diet_df[['business_id', 'attributes.DietaryRestrictions']]
business_diet_df
```

```python
# get list of all possible categories
list(business_diet_df['attributes.DietaryRestrictions'])[0]
```

```python
business_diet_df['attributes.DietaryRestrictions'] = business_diet_df['attributes.DietaryRestrictions'].str.replace("\'", "\"").str.replace("False", "\"False\"").str.replace("True", "\"True\"")
```

```python
diets_df = business_diet_df.join(business_diet_df['attributes.DietaryRestrictions'].apply(json.loads).apply(pd.Series))
```

### Vegetarian
